```python
from pandas import DataFrame as df
import numpy as np


def find_min_distance( matrix):
        min_distance = float('inf')
        min_i, min_j = -1, -1
        n=len(matrix[0])
        for i in range(n):
            for j in range(i + 1, n):
                if (matrix[i][j] < min_distance) and ( i !=j):
                    min_distance = matrix[i][j]
                    min_i, min_j = i, j
        return min_i, min_j


def UPGMA_Algorithm(distance_matrix, letters):
    # initial matrix
    Edges = [[i] for i in range(len(letters))]
    Tree_Guid = []
    print("Distance matrix")
    show=lambda a,b:df(a,index=b,columns=b)
    print(show(distance_matrix,letters))
    print("#"*60)
    while len(letters) > 1:
        
        Min_args =find_min_distance(distance_matrix)
        
        Min_value = distance_matrix[Min_args[0], Min_args[1]]
        
        new_Edge = f"{letters[Min_args[1]]}-{letters[Min_args[0]]}"
        
        Tree_Guid.append(f" [ {letters[Min_args[1]]} ] =>  {Min_value / 2} <= [ {letters[Min_args[0]]} ] -")
        
        new_value = (distance_matrix[Min_args[0]] + distance_matrix[Min_args[1]]) / 2

        distance_matrix[Min_args[0]] = new_value
        distance_matrix[:, Min_args[0]] = new_value
        
        np.fill_diagonal(distance_matrix, 0)
        
        hide = np.ones(len(distance_matrix), dtype=bool)
        hide[Min_args[1]] = False
        distance_matrix = distance_matrix[hide][:, hide]
        
        np.fill_diagonal(distance_matrix, 0)
        
        print("=" * 30)
        letters[Min_args[0]] = new_Edge

        
        del letters[Min_args[1]]

        
        print("\nTree guid :")
        for label in Tree_Guid:
            print(label)
        print("Distance matrix")
        print(show(distance_matrix,letters))

    

# Example usage
Letters = ["A", "B", "C", "D"]
distance_matrix = np.array([[0, 17.0, 21.0, 27.0],
                            [17.0, 0, 12.0, 18.0],
                            [21.0, 12.0, 0, 14.0],
                            [27.0, 18.0, 14.0, 0]])

upgma_result = UPGMA_Algorithm(distance_matrix, Letters)


#=======================================================================================================
```
